## Introduction

The agent has a map of the maze it is in and the environment is assumed to be **deterministic, discrete, and known.** The agent must use the map to plan a path through the maze from the starting location $S$ to the goal location $G$.  This is a planing exercise for a goal-based agent, so you do not need to implement an environment, just use the map to search for a path. Once the plan is made, the agent in a deterministic environment (i.e., the transition function is deterministic with the outcome of each state/action pair fixed and no randomness) can just follow the path and does not need to care about the percepts.
This is also called an **[open-loop system](https://en.wikipedia.org/wiki/Open-loop_controller).**
The execution phase is trivial and we do not implement it in this exercise.

Tree search algorithm implementations that you find online and used in general algorithms courses have often a different aim. These algorithms assume that you already have a tree in memory. We are interested in dynamically creating a search tree with the aim of finding a good/the best path from the root noteto the goal state. Follow the pseudo code presented in the text book (and replicated in the slides) closely. Ideally, we would like to search only a small part of the maze, i.e., create a search tree with as few nodes as possible.

## Tree structure

Here is an implementation of the basic node structure for the search algorithms (see Fig 3.7 on page 73). I have added a method that extracts the path from the root node to the current node. It can be used to get the path when the search is completed.

```python
class Node:
    def __init__(self, pos, parent, action, cost):
        self.pos = tuple(pos)    # the state; positions are (row,col)
        self.parent = parent     # reference to parent node. None means root node.
        self.action = action     # action used in the transition function (root node has None)
        self.cost = cost         # for uniform cost this is the depth. It is also g(n) for A* search

    def __str__(self):
        return f"Node - pos = {self.pos}; action = {self.action}; cost = {self.cost}"
    
    def __eq__(self, other):
        if self.pos==other.pos:
            return True
        else:
            return False

    def get_path_from_root(self):
        """returns nodes on the path from the root to the current node."""
        node = self
        path = [node]
    
        while not node.parent is None:
            node = node.parent
            path.append(node)
        
        path.reverse()
        
        return(path)
```

If needed, then you can add more fields to the class like the heuristic value $h(n)$ or $f(n)$.

## Task 2: Uninformed search: Breadth-first and depth-first [40 Points]

Implement these search strategies. Follow the pseudocode in the textbook/slides. You can use the tree structure shown above to extract the final path from your solution.

__Notes:__
* You can find maze solving implementations online that use the map to store information. While this is an effective idea for this two-dimensional navigation problem, it typically cannot be used for other search problems. Therefore, follow the textbook and only store information in the tree created during search, and use the `reached` and `frontier` data structures.
* DSF can be implemented using the BFS tree search algorithm and simply changing the order in which the frontier is expanded (this is equivalent to best-first search with path length as the criterion to expand the next node). However, to take advantage of the significantly smaller memory footprint of DFS, you need to implement DFS in a different way without a `reached` data structure and by releasing the memory for nodes that are not needed anymore. 
* If DFS does not use a `reached` data structure, then its cycle checking abilities are limited. Remember, that DSF is incomplete if cycles cannot be prevented. You will see in your experiments that open spaces are a problem.

```python
# expand function to explore possible next nodes
def Expand(maze, node):
    x= node.pos[0]
    y= node.pos[1]
    next_node=[]
    # check east
    if mh.look(maze, (x+1, y)) != "X":
        next_node.append(Node(pos=(x+1,y), parent=node, action="East", cost=node.cost+1))
    # check west
    if mh.look(maze, (x-1, y)) != "X":
        next_node.append(Node(pos=(x-1,y), parent=node, action="West", cost=node.cost+1))
    # check north
    if mh.look(maze, (x, y+1)) != "X":
        next_node.append(Node(pos=(x,y+1), parent=node, action="North", cost=node.cost+1))
    # check south
    if mh.look(maze, (x, y-1)) != "X":
        next_node.append(Node(pos=(x,y-1), parent=node, action="South", cost=node.cost+1))
    return next_node


def bfs(maze):
    c=0 # limit of steps counter
    visited=[]  # reached
    queue=[]  # frontier
    max_depth=0  
    max_frontier=0
    # set "start" as the root of our tree
    root=Node(pos=mh.find_pos(maze, what="S"), parent=None, action=None, cost=0)
    # check if endpoint G is reached
    if mh.look(maze, root.pos) != "G":  # if endpoint is not reached at the very beginning
        queue.append(root)  # push root to frontier
        visited.append(root) # push root to reached
        
        while queue:
            # get max frontier size
            if len(queue) > max_frontier:
                max_frontier = len(queue)
            # get current node from front of queue
            curr = queue.pop(0)
            # update tree depth
            tree_depth=len(curr.get_path_from_root())
            if tree_depth > max_depth:
                max_depth =tree_depth


            # find where to go next
            next_nodes= Expand(maze, curr)
            # print(next_nodes)
            for node in next_nodes:
                # if endpoint is reached
                if mh.look(maze, node.pos)=="G": 
                    return (node, visited, len(visited), max_depth, len(visited)+max_depth, max_frontier)
                if node not in visited:
                    # print(node.pos)
                    visited.append(node)
                    queue.append(node)
                    
            # counter++
            # print(c)
            c = c+1

        return "can't find route"

    else: 
        return root  # return root if endpoint is at the root





def dfs(maze):
    c=0 # limit of steps counter
    visited=[]  # reached
    stack=[]  # frontier
    max_depth=0  
    max_frontier=0 
    # set "start" as the root of our tree
    root=Node(pos=mh.find_pos(maze, what="S"), parent=None, action=None, cost=0)
    # check if endpoint G is reached
    if mh.look(maze, root.pos) != "G":  # if endpoint is not reached at the very beginning
        stack.append(root)  # push root to frontier
        visited.append(root) # push root to reached
        
        while stack and c<1500:

            # get max frontier size
            if len(stack) > max_frontier:
                max_frontier = len(stack)
            # get current node from front of stack
            curr = stack.pop(len(stack)-1)
            # update tree depth
            tree_depth=len(curr.get_path_from_root())
            if tree_depth > max_depth:
                max_depth =tree_depth
            
            # find where to go next
            next_nodes= Expand(maze, curr)
            # print(next_nodes)
            for node in next_nodes:
                # if endpoint is reached
                if mh.look(maze, node.pos)=="G": 
                    return (node, visited, len(visited), max_depth, len(visited)+max_depth, max_frontier)
                path = curr.get_path_from_root()
                
                if node not in path:
                    # print(node.pos)
                    visited.append(node)
                    stack.append(node)
                    
            # counter++
            # print(c)
            c = c+1
        # when ran to max
        return (node, visited, len(visited), max_depth, len(visited)+max_depth, max_frontier)

    else: 
        return root  # return root if endpoint is at the root



        
```

How does BFS and DFS deal with loops (cycles)?

While BFS searches through the "visited" array that covers all the past nodes to avoid repetition/looping, this version of DFS only check for duplicate in the current path taken. That makes DFS more vulnerable to loops.

Are your implementations complete and optimal? Explain why. What is the time and space complexity of each of **your** implementations?

#### BFS: 
* Complete: It is complete. 
* Optimal: It uses uniform-cost search, therefore optimal. 
* Time: sum of the number of nodes created at each depth level in a tree of depth d: $1+b+b^2+...+b^d= O(b^2)$
* Space complexity: $O(b^d)$

#### DFS: 
* Complete: Only would be complete only in finite search spaces. Might not negate all the chances to run into cycles.  
* Optimal: No, because it returns the first solutions that comes up 
* Time: 
b: max branching factor, m: max depth of tree, $O(b^m)$
* Space complexity: $O(b^m)$

## Task 3: Informed search: Implement greedy best-first search and A* search  [20 Points]

You can use the map to estimate the distance from your current position to the goal using the Manhattan distance (see https://en.wikipedia.org/wiki/Taxicab_geometry) as a heuristic function. Both algorithms are based on Best-First search which requires only a small change from the BFS algorithm you have already implemented (see textbook/slides). 

```python

def greedy(maze):
    c=0 # limit of steps counter
    visited=[]  # reached
    queue=[]  # frontier
    max_depth=0  
    max_frontier=0  
    # set "start" as the root of our tree
    root=Node(pos=mh.find_pos(maze, what="S"), parent=None, action=None, cost=0)
    # check if endpoint G is reached
    if mh.look(maze, root.pos) != "G":  # if endpoint is not reached at the very beginning
        queue.append(root)  # push root to frontier
        visited.append(root) # push root to reached
        
        while queue:
            # get max frontier size
            if len(queue) > max_frontier:
                max_frontier = len(queue)
            # get current node from front of queue
            curr = queue.pop(0)
            # update tree depth
            tree_depth=len(curr.get_path_from_root())
            if tree_depth > max_depth:
                max_depth =tree_depth
            
            # find where to go next
            next_nodes= Expand(maze, curr)
            # print(next_nodes)
            for node in next_nodes:
                # if endpoint is reached
                if mh.look(maze, node.pos)=="G": 
                    return (node, visited, len(visited), max_depth, len(visited)+max_depth, max_frontier)
                if node not in visited:
                    # print(node.pos)
                    visited.append(node)
                    queue.append(node)

            queue.sort(key = lambda x : abs(x.pos[1]-gy)+abs(x.pos[0]-gx))     
            # counter++
            # print(c)
            c = c+1

        return "can't find route"

    else: 
        return root  # return root if endpoint is at the root



def Astar(maze):
    c=0 # limit of steps counter
    visited=[]  # reached
    queue=[]  # frontier
    max_depth=0  
    max_frontier=0  
    # set "start" as the root of our tree
    root=Node(pos=mh.find_pos(maze, what="S"), parent=None, action=None, cost=0)
    # check if endpoint G is reached
    if mh.look(maze, root.pos) != "G":  # if endpoint is not reached at the very beginning
        queue.append(root)  # push root to frontier
        visited.append(root) # push root to reached
        
        while queue:
            # get max frontier size
            if len(queue) > max_frontier:
                max_frontier = len(queue)
            # get current node from front of queue
            curr = queue.pop(0)
            # update tree depth
            tree_depth=len(curr.get_path_from_root())
            if tree_depth > max_depth:
                max_depth =tree_depth
            
            # find where to go next
            next_nodes= Expand(maze, curr)
            # print(next_nodes)
            for node in next_nodes:
                # if endpoint is reached
                if mh.look(maze, node.pos)=="G": 
                    return (node, visited, len(visited), max_depth, len(visited)+max_depth, max_frontier)
                if node not in visited:
                    # print(node.pos)
                    visited.append(node)
                    queue.append(node)

            queue.sort(key = lambda x : abs(x.pos[1]-gy)+abs(x.pos[0]-gx)+ x.cost)     
            # counter++
            # print(c)
            c = c+1

        return "can't find route"

    else: 
        return root  # return root if endpoint is at the root
```

Are your implementations complete and optimal? What is the time and space complexity?

#### GBS: 
* Complete: It is complete. 
* Optimal: No.
* Time (worst case): $O(b^m)$, b: max branching factor, m: max depth of tree
* Time (best case): $O(b*m)$, if h(n) is accurate
* Space complexity: $O(b^d)$ or $O(b*m)$

#### A*: 
* Complete: Yes
* Optimal: Yes
* Time: # of nodes for $ f(n) <= C*$
* Space complexity: # of nodes for $ f(n) <= C*$

Discuss the most important lessons you have learned from implementing the different search strategies. 

* Overall

A* and greedy are the more efficient algorithm at searching for a path among 4 of the implemented methods if we judge this "efficiency" based on the 4 attributes we listed above. While GBS outperforms A* on some aspects (max num of nodes in memory and number of nodes expanded), we would still prefer A* as it outperforms greedy in general, and consistently gave us more optimal/low-cost pathings.

* BFS vs DFS

While DFS could cut some corners on memory usage by using "path taken" instead of searching through the "all paths attempted" array, it limits its ability to efficiently traverse through loops and open areas, as DFS would attept to make the "old mistakes" and get stuck in a loop.
